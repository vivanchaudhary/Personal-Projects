# Section 2: Exploratory Data Analysis

In order to give a comprehensive picture of a dataset and provide insights into its size, shape, central tendency, variability, and distribution, descriptive statistics are essential. Key points that descriptive statistics can shed light on are as follows:

1) The dataset's size:

    Meaning: Indicates how many observations or data points are included in the dataset.
    Measure: The dataset's length or count.

2) Form of the Distribution:

    Meaning: Indicates the data's distributional properties.
    Skewness and kurtosis are the measures. A distribution that is left-skewed is indicated by negative skewness, and a distribution that is right-skewed by positive skewness. Kurtosis calculates the distribution's tail heaviness.

3) Central Propensity

    Meaning: Determines the average or centre value that the data points cluster around.
    Calculations: Mode, median, and mean. While the median is more resilient in the face of extreme numbers, the mean is more susceptible to outliers.

4) Changeability:

    Signifies the extent of dispersion or spread within the dataset.
    Measures include interquartile range (IQR), variance, range, and standard deviation. Greater variability is indicated by higher values.

# Section 4: Data Science

One question that arises regarding the ICC World Cup matches data set is that is there any correlation between how teams are chosen to compete against each other. For this we can look at the correltion between the team IDs of both the teams playing in the match. We can further answer this question with the help of data science!

## Section 4.1: Making a numerical dataset of Team IDs

```python
numerical_cols = ['id','venueId','team1Id','team2Id']
df_numerical = df[numerical_cols]
df_numerical
```

Now that we have the numerical columns let us do the correlation.

```python
df_numerical.corr()
```

A correlation of +0.683824 between team1id and team2id shows that there is a strong correlation between them. That means when once increase the other also increases. This means that teams with higher teamIDs tend to compete against other teams with higher IDs and teams with lower IDs tend to compete against other teams with lower IDs.

We can also take a look at this with the help of a scatter plot below. 

```python
df_numerical.plot.scatter(x='team1Id' , y='team2Id')
```
